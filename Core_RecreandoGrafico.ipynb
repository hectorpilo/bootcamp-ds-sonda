# **Reproducir un Gráfico Completo**

deberán reproducir un gráfico completo de Seaborn basado en una imagen del gráfico terminado. El gráfico incluirá elementos como cuadrículas, leyendas, anotaciones y flechas.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset
df = sns.load_dataset("tips")

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', ax=ax)

# Título y etiquetas
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.grid(True)

# Encontrar el punto con el tip más alto
max_tip_idx = df['tip'].idxmax()
x_max = df.loc[max_tip_idx, 'total_bill']
y_max = df.loc[max_tip_idx, 'tip']

# Añadir anotación con flecha negra
ax.annotate('Mayor tip en la cena',
            xy=(x_max, y_max),
            xytext=(x_max  -6, y_max + 5),
            arrowprops=dict(arrowstyle='->', color='black', linewidth=2),
            fontsize=12, color='black')


# Añadir flecha verde desde un punto hacia el punto más alto
ax.arrow(30, 7, x_max - 30, y_max - 7,
         head_width=0.8, head_length=0.5, fc='green', ec='green')

# Leyenda en esquina superior derecha
ax.legend(loc='upper right', fontsize=10)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```
