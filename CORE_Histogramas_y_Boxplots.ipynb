```python
df2017 = df2017.rename(columns={'Happiness.Score': 'Score'})

import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Diccionario con los DataFrames por año
dataframes = {
    2015: df2015,
    2016: df2016,
    2017: df2017,
    2018: df2018,
    2019: df2019
}

# Paso 2: Función para renombrar columnas y graficar
def procesar_y_graficar(df, año):
    # Renombramos columnas para que todos tengan los mismos nombres clave
    df = df.rename(columns={
        'Happiness Score': 'Score',
        'Score': 'Score',
        'Economy (GDP per Capita)': 'GDP per capita',
        'Economy..GDP.per.Capita.': 'GDP per capita',
        'GDP per capita': 'GDP per capita',
        'Health (Life Expectancy)': 'Healthy life expectancy',
        'Health..Life.Expectancy.': 'Healthy life expectancy',
        'Healthy life expectancy': 'Healthy life expectancy',
        'Freedom': 'Freedom to make life choices',
        'Freedom to make life choices': 'Freedom to make life choices'
    })

    # Crear figura de 2x2
    fig, axs = plt.subplots(2, 2, figsize=(12, 8))
    fig.suptitle(f"Distribución de Indicadores de Felicidad - {año}", fontsize=14)

    # Variables a graficar
    variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']
    posiciones = [(0,0), (0,1), (1,0), (1,1)]

    for var, pos in zip(variables, posiciones):
        if var in df.columns:
            axs[pos].hist(df[var], bins=15, color='skyblue', edgecolor='black')
            axs[pos].set_title(var)
            axs[pos].set_xlabel("Valor")
            axs[pos].set_ylabel("Frecuencia")
        else:
            axs[pos].text(0.5, 0.5, 'No disponible', ha='center', va='center')
            axs[pos].set_title(var)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Paso 3: Ejecutamos la función para cada año
for año, df in dataframes.items():
    procesar_y_graficar(df, año)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Librería ideal para boxplots

# Asegúrate de tener seaborn instalado: pip install seaborn

# Diccionario de DataFrames
dataframes = {
    2015: df2015,
    2016: df2016,
    2017: df2017,
    2018: df2018,
    2019: df2019
}

# Variables a graficar
variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

# Función para renombrar columnas de forma uniforme
def estandarizar(df):
    return df.rename(columns={
        'Happiness Score': 'Score',
        'Score': 'Score',
        'Economy (GDP per Capita)': 'GDP per capita',
        'Economy..GDP.per.Capita.': 'GDP per capita',
        'Health (Life Expectancy)': 'Healthy life expectancy',
        'Health..Life.Expectancy.': 'Healthy life expectancy',
        'Freedom': 'Freedom to make life choices',
        'Freedom to make life choices': 'Freedom to make life choices',
        'Country or region': 'Country',
        'Overall rank': 'Happiness Rank'
    })

# Unimos todos los datos en un solo DataFrame
datos_combinados = pd.DataFrame()

for año, df in dataframes.items():
    df = estandarizar(df)
    df['Año'] = año  # agregamos la columna "Año"
    datos_combinados = pd.concat([datos_combinados, df], ignore_index=True)

# Graficamos boxplots por variable
for variable in variables:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Año', y=variable, data=datos_combinados, palette='pastel')
    plt.title(f'Distribución de "{variable}" entre 2015 y 2019')
    plt.xlabel("Año")
    plt.ylabel(variable)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

# **Graficos Combinados**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Variables a analizar
variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

# Diccionario para guardar las figuras si luego deseas exportarlas
figuras = {}

# Estilo de seaborn
sns.set(style="whitegrid")

for variable in variables:
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Histograma con curva KDE
    sns.histplot(
        data=datos_combinados,
        x=variable,
        kde=True,
        ax=axs[0],
        color='skyblue',
        edgecolor='black',
        alpha=0.8,
        linewidth=1.2
    )
    axs[0].set_title(f'Histograma de {variable}', fontsize=13)
    axs[0].set_xlabel(variable)
    axs[0].set_ylabel('Frecuencia')

    # Boxplot sin usar hue ni palette para evitar warning
    sns.boxplot(
        data=datos_combinados,
        x='Año',
        y=variable,
        ax=axs[1],
        linewidth=1.2
    )
    axs[1].set_title(f'Boxplot de {variable} por Año', fontsize=13)
    axs[1].set_xlabel('Año')
    axs[1].set_ylabel(variable)

    # Título general
    plt.suptitle(f'Análisis combinado: {variable}', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])

    # Guardamos por si se quiere exportar
    figuras[variable] = fig

plt.show()
```

# **A modo de juego, realicé analisis de la información recopilada en chile respecto a la felicidad entre el año 2015 y 2019.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Diccionario con tus DataFrames por año
dataframes = {
    2015: df2015,
    2016: df2016,
    2017: df2017,
    2018: df2018,
    2019: df2019
}

# Función para renombrar columnas de forma uniforme
def estandarizar(df):
    df = df.rename(columns={
        'Happiness Score': 'Score',
        'Score': 'Score',
        'Economy (GDP per Capita)': 'GDP per capita',
        'Economy..GDP.per.Capita.': 'GDP per capita',
        'GDP per capita': 'GDP per capita',
        'Health (Life Expectancy)': 'Healthy life expectancy',
        'Health..Life.Expectancy.': 'Healthy life expectancy',
        'Healthy life expectancy': 'Healthy life expectancy',
        'Freedom': 'Freedom to make life choices',
        'Freedom to make life choices': 'Freedom to make life choices',
        'Country or region': 'Country',
        'Country': 'Country',
        'Overall rank': 'Happiness Rank'
    })
    return df

# Elegimos el país y la variable a analizar
pais = 'Chile'
variable = 'Score'

# Lista para guardar valores año a año
resultados = []

# Recorremos los DataFrames por año
for año, df in dataframes.items():
    df = estandarizar(df)  # Renombramos las columnas
    if variable in df.columns and 'Country' in df.columns:
        fila = df[df['Country'] == pais]
        if not fila.empty:
            valor = fila[variable].values[0]
            resultados.append((año, valor))
        else:
            print(f"⚠️ País '{pais}' no encontrado en {año}")
    else:
        print(f"❌ Columna '{variable}' o 'Country' no está en el DataFrame del {año}")

# Creamos el DataFrame final con los resultados
df_pais = pd.DataFrame(resultados, columns=['Año', variable])

# Verificamos si hay datos para graficar
if not df_pais.empty:
    # Gráfico de evolución
    plt.figure(figsize=(10, 5))
    plt.plot(df_pais['Año'], df_pais[variable], marker='o', color='blue', linestyle='-', linewidth=2)
    plt.title(f"Evolución de '{variable}' en {pais} (2015-2019)", fontsize=14)
    plt.xlabel('Año')
    plt.ylabel(variable)
    plt.grid(True)
    plt.xticks(df_pais['Año'])  # Mostrar todos los años como etiquetas
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No hay datos disponibles para graficar.")
```
